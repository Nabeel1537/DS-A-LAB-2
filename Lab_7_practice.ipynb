```python
#remove the nan values
pokemon.dropna(axis=0, inplace = True)
pokemon.isnull().sum()
pokemon.info()
```

```python
pokemon['generation_id'].value_counts()
```

```python
#making the bar chart
#generation_id is catagorical
sb.countplot(data=pokemon, x='generation_id');

gen_order=pokemon['generation_id'].value_counts().index
```

```python
#for choosing the color of the bar chart
#lets use one color
sb.color_palette()
```

```python
base_color=sb.color_palette()[2]
```

```python
sb.countplot(data=pokemon, x='generation_id', color=base_color, order=gen_order)
```

```python
sb.countplot(data=pokemon, x='type_1');
base_color=sb.color_palette()[2]
plt.xticks(rotation=90)
plt.show()
```

```python
sb.countplot(data=pokemon, y='type_1');
```

```python
type_order=pokemon['type_1'].value_counts().index
sb.countplot(data=pokemon, y='type_1', color=base_color,order=type_order)
```
